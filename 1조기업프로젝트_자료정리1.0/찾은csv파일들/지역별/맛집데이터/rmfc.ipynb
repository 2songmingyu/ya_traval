```python
import csv
import os
import tempfile
import shutil

def remove_rows_with_strings_in_folder(folder_path, target_strings):
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            temp_file_path = os.path.join(tempfile.gettempdir(), 'temp.csv')
            rows_to_keep = []
            
            # 파일 인코딩을 'utf-8'로 시도하고, 에러가 발생하면 'cp949'로 시도합니다
            try:
                with open(file_path, 'r', encoding='utf-8-sig') as csv_file, \
                     open(temp_file_path, 'w', newline='', encoding='utf-8-sig') as temp_file:
                    csv_reader = csv.reader(csv_file)
                    csv_writer = csv.writer(temp_file)
                    
                    for row in csv_reader:
                        if not any(target_string in cell for target_string in target_strings for cell in row):
                            rows_to_keep.append(row)
                            csv_writer.writerow(row)
            
            except UnicodeDecodeError:
                with open(file_path, 'r', encoding='cp949') as csv_file, \
                     open(temp_file_path, 'w', newline='', encoding='cp949') as temp_file:
                    csv_reader = csv.reader(csv_file)
                    csv_writer = csv.writer(temp_file)
                    
                    for row in csv_reader:
                        if not any(target_string in cell for target_string in target_strings for cell in row):
                            rows_to_keep.append(row)
                            csv_writer.writerow(row)
            
            shutil.move(temp_file_path, file_path)  # 임시 파일을 원본 파일로 이동
            
            print(f"Processed file: {filename}")
            for row in rows_to_keep:
                print(row)
            print()

folder_path = 'C:/ya_traval/1조기업프로젝트_자료정리1.0/찾은csv파일들/지역별/맛집데이터/인천'
target_strings = ['DT', '스타벅스', '맥도날드', '버거킹', '투썸플레이스', '쿠우쿠우', '뚜레쥬르', '파리바게뜨', '파리크라상', '아웃백', 
                  'SUBWAY', 'VIPS', '이디야', '엔제리너스', '카페베네', '탐앤탐스', '롯데리아', 'KFC', 'BHC', '던킨', '명륜진사갈비',
                  '도미노피자', '피자헛', '파스쿠찌', '교촌치킨', '폴바셋', '맘스터치', '김밥천국', '애슐리']

remove_rows_with_strings_in_folder(folder_path, target_strings)
```
