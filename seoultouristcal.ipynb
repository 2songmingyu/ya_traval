```python
# 인구 비율이 있는 엑셀 파일의 경로
entertouristratio = '1조기업프로젝트_자료정리1.0/찾은csv파일들/지역별/주요관광지성연령별/서울.csv'
```

```python
# 입장객 합계 엑셀 파일 로드
df_entertourist = pd.read_excel(entertouristall)

# 인구 비율 csv 파일 로드
df_ratio = pd.read_csv(entertouristratio)
```

```python
# 필요없는 여러 개의 컬럼 제거
columns_to_drop = ['Unnamed: 0', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20']
df_ratio = df_ratio.drop(columns_to_drop, axis=1)
print(df_ratio)
```

```python
# 내/외국인 칼럼 이름
nationalitycol = '내/외국인'

# 내/외국인 칼럼에서 합계 행만 남기기
df_entertourist = df_entertourist[df_entertourist[nationalitycol] == '합계']
```

```python
df_entertourist.drop(columns=['총계(2022.01 ~ 2022.12)'], inplace=True) # 서울 제외 '(2022년 ~ 2022년)', 서울 '총계(2022.01 ~ 2022.12)'
```

```python
# 입장객 합계 계산
cs_entertourist = ['관광지', '2022년'] # 서울 제외 - '총계', 서울 - '2022년'
df_touristsum = df_entertourist[cs_entertourist]
```

```python
from fuzzywuzzy import process

# 첫 번째 데이터프레임의 행 데이터와 두 번째 데이터프레임의 행 데이터를 비교하여 가장 유사한 행을 찾습니다.
threshold_similarity = 80  # 필요에 따라 유사도 임계값을 조정합니다.

selected_rows = []
for index, row1 in df_touristsum.iterrows():
    name1 = row1['관광지']  # 첫 번째 데이터프레임에서 이름에 해당하는 컬럼을 지정해야 합니다.
    
    best_match = process.extractOne(name1, df_ratio['관광지'])[0]  # 두 번째 데이터프레임에서 이름에 해당하는 열 이름을 지정합니다.
    similarity_score = process.extractOne(name1, df_ratio['관광지'])[1]
    if similarity_score >= threshold_similarity:
        selected_rows.append((row1))

# 선택된 행 데이터로 새로운 데이터프레임을 생성합니다.
new_df = pd.DataFrame(selected_rows, columns=df_touristsum.columns)  # 필요에 따라 열 이름을 수정합니다.

# 선택된 행 데이터를 출력합니다.
print(new_df)
```

```python
# 새로운 데이터프레임과 세 번째 데이터프레임을 비교하고 병합합니다.
merged_df = pd.merge(new_df, df_ratio, on='관광지', how='inner')

# 선택된 데이터를 출력합니다.
print(merged_df)
```

```python
pd.DataFrame(merged_df)
```

```python
# 총계 컬럼을 활용하여 관광객 숫자 계산
columns_to_calculate = ['10대 이하(남)', '10대 이하(여)', '20대 (남)', '20대 (여)', '30대 (남)', '30대 (여)', '40대 (남)', '40대 (여)', '50대 (남)', '50대 (여)', '60대 (남)', '60대 (여)', '70대 이상 (남)', '70대 이상 (여)']
for column in columns_to_calculate:
    tourist_count_column_name = column + ' 관광객 수'
    merged_df[tourist_count_column_name] = (merged_df[column] / 100) * merged_df['2022년'] # 서울 제외 '총계', 서울만 '2022년'
    merged_df[tourist_count_column_name] = merged_df[tourist_count_column_name].astype(int)

# '관광객 수' 컬럼이 없는 컬럼들 삭제
columns_to_drop = ['2022년'] + columns_to_calculate # 서울 제외 '총계', 서울만 '2022년'
merged_df.drop(columns=columns_to_drop, inplace=True)

# 결과 출력
print(merged_df)
```
